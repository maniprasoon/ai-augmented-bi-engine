```python
import pandas as pd

df = pd.read_csv("../outputs/master_table.csv")
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
monthly = df.groupby(
    df['order_date'].dt.to_period('M')
).agg({
    'revenue': 'sum',
    'order_id': 'nunique',
    'delivery_days': 'mean'
}).reset_index()
```

```python
total_revenue = df['revenue'].sum()
avg_order_value = total_revenue / df['order_id'].nunique()

monthly['revenue_growth_pct'] = monthly['revenue'].pct_change() * 100
```

```python
kpis = pd.DataFrame({
    "metric": [
        "Total Revenue",
        "Average Order Value",
        "Latest Monthly Revenue Growth (%)",
        "Avg Delivery Days"
    ],
    "value": [
        total_revenue,
        avg_order_value,
        monthly['revenue_growth_pct'].iloc[-1],
        df['delivery_days'].mean()
    ]
})
```

```python
kpis.to_csv("../outputs/kpis.csv", index=False)
```
